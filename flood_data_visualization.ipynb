```python
import pandas as pd

# Load CSV file into a pandas DataFrame
water_bodies = pd.read_csv('Hydro.csv')

# Print the first 5 rows of the DataFrame
# drop all columns except for the ones we need
water_bodies = water_bodies['the_geom']
water_bodies.head()
```

```python
import numpy as np
water_coordinates = []
for i in water_bodies.values:
    stringo = i
    # remove the word 'MULTIPOLYGON' from the string and ((())) from the string
    stringo = stringo.replace('MULTIPOLYGON (((', '')
    stringo = stringo.replace('(', '')
    stringo = stringo.replace(')', '')
    # separate the string by commas
    locations = stringo.split(',')
    # create a new list to store the coordinates
    coordinates = []
    # loop through the locations and separate the coordinates
    for location in locations:
        # separate the coordinates by spaces
        coordinate = location.strip().split(' ')
        if len(coordinate) != 2:
            continue
        if not (-87.85 < float(coordinate[0]) and float(coordinate[0]) < -87):
            continue
        if not (float(coordinate[1]) < 42.2 and float(coordinate[1]) >35):
            continue
        # append the coordinates to the coordinates list
        coordinates.append(coordinate)
    water_coordinates.append(coordinates)
print(water_coordinates[0])
```

```python
import matplotlib.pyplot as plt

# create empty lists to store the coordinates
longitudes_water = []
latitudes_water = []

# loop through the water_coordinates list and extract the coordinates
for coordinates in water_coordinates:
    for coordinate in coordinates:
        longitudes_water.append(float(coordinate[0]))
        latitudes_water.append(float(coordinate[1]))
print(max(longitudes_water))
print(min(longitudes_water))
# increase the size of the plot
plt.figure(figsize=(20, 10))

# plot the coordinates in a scatter plot
plt.scatter(longitudes_water, latitudes_water, c='blue', s=0.1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Zip Code Coordinates')
plt.show()
```

```python
flood_claims = pd.read_csv('Claims_flooding_zipcode_chicago.csv')
flood_claims = flood_claims[['Zip code', 'Total claims']]
flood_claims.head()
```

```python
zip_codes = pd.read_csv('Zip_codes.csv')
zip_codes = zip_codes[['the_geom', 'ZIP']]
zip_codes.head()
```

```python
# loop through flood_claims zip_code, and query zip_codes in zip_code data_frame and fetch the_geom
zip_codes_geom = []
for zip_code in flood_claims['Zip code']:
    zip_codes_geom.append((zip_codes[zip_codes['ZIP'] == zip_code]['the_geom'].values, zip_code, 
                        flood_claims[flood_claims['Zip code'] == zip_code]['Total claims'].values))
# convert zip_codes_geom to a data_frame
zip_codes_geom = pd.DataFrame(zip_codes_geom)

zip_codes_geom.head()
# drop fist two rows 
zip_codes_geom = zip_codes_geom.drop([0,1])
# print all the zip_codes_geom where 0 is empty
print(zip_codes_geom[zip_codes_geom[0].apply(lambda x: len(x) == 0)])
# sort the zip_codes_geom by the number of claims
zip_codes_geom = zip_codes_geom.sort_values(by=[2], ascending=False)
# print the top 5 zip_codes_geom
zip_codes_geom.head()
```

```python
import numpy as np
zip_to_coordinates = {}
for i in zip_codes_geom.values:
    stringo = i[0][0]
    # remove the word 'MULTIPOLYGON' from the string and ((())) from the string
    stringo = stringo.replace('MULTIPOLYGON (((', '')
    stringo = stringo.replace('(', '')
    stringo = stringo.replace(')', '')
    # separate the string by commas
    locations = stringo.split(',')
    # create a new list to store the coordinates
    coordinates = []
    # loop through the locations and separate the coordinates
    for location in locations:
        # separate the coordinates by spaces
        coordinate = location.strip().split(' ')
        # append the coordinates to the coordinates list
        coordinates.append(coordinate)
        zip_to_coordinates[i[1]] = coordinates
print(zip_to_coordinates.keys())
print(zip_to_coordinates[60628])
```

```python
import matplotlib.pyplot as plt

# create empty lists to store the coordinates
longitudes = []
latitudes = []

long_mean = []
lat_mean = []

# loop through the zip_to_coordinates dictionary and extract the coordinates
for coordinates in zip_to_coordinates.values():
    # take the mean of the coordinates and append them to the lists
    long_mean.append(np.mean([float(coordinate[0]) for coordinate in coordinates]))
    lat_mean.append(np.mean([float(coordinate[1]) for coordinate in coordinates]))
        
    for coordinate in coordinates:
        longitudes.append(float(coordinate[0]))
        latitudes.append(float(coordinate[1]))

# increase the size of the plot
plt.figure(figsize=(20, 10))

# plot the coordinates in a scatter plot
plt.scatter(longitudes, latitudes, c='k', s=0.1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Zip Code Coordinates')
# the long_mean and lat_mean are the x, y, the color is the number of claims
plt.scatter(long_mean, lat_mean, c=zip_codes_geom[2], s=100, cmap='plasma')
plt.colorbar(label='Number of Claims', location='left')

plt.scatter(longitudes_water, latitudes_water, c='blue', s=0.1)
plt.scatter(longitudes_green, latitudes_green, c='green', s=0.1)
# annotate the zip codes with the number of claims
for i in range(len(long_mean)):
    # make the annotation bold
    plt.annotate(zip_codes_geom[2].values[i][0], (long_mean[i], lat_mean[i]), weight='bold')
    # highlight the zip code with the most claims
    if i == 0:
        plt.annotate(zip_codes_geom[2].values[i][0], (long_mean[i], lat_mean[i]), weight='bold', color='red')
        print(long_mean[i])
# add the elevation data to the plot
elevation.plot.scatter(x='LONGITUDE', y='LATITUDE', c='ELEVATION', 
                       colormap='YlGnBu', s=500, ax=plt.gca(), zorder=0)
plt.show()
```
